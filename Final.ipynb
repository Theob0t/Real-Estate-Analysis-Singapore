## Moving Average as Prediction
Compute the mean on the historical data and predict the value for the next data

The moving average value can also be used directly to make predictions

As new observations are made available (e.g. monthly), the model can be updated and a prediction made for the next month.

```python
# prepare problem
X = prices.values
# start with 3 past months
window = 3
train = [X[i] for i in range(window)]
test = [X[i] for i in range(window, len(X))]
predictions = []

# walk forward over time steps in test
for t in range(len(test)):
    # train increase of 1 each month (adding the value of the previous month)
    length = len(train)
    # pred value of the month by computing the mean of all the previous month
    pred = np.mean([train[i] for i in range(length-window, length)])

    obs = test[t]

    predictions.append(pred)
    train.append(obs)

rmse = np.sqrt(mean_squared_error(test, predictions))
abs_err = mean_absolute_error(test, predictions)
print('RMSE: %.2f' % rmse)
print('Abs error: %.2f' % abs_err)
```

```python
# zoomed plot of predictions
plt.plot(test[50:150])
plt.plot(predictions[50:150], color='red')
plt.title('Zoomed Plot of Moving Average Predictions')
plt.show()
```

Using the Moving Average method (computing the average sqm price of all previous months), we can predict the sqm price of the next month with an accuracy of $50.

## Stationarity in Time Series Data

Our time series dataset may contain a trend. A trend is a continued increase or decrease in the series over time

A time series where the seasonal component has been removed is called seasonal stationary. If a dataset does not have a trend or we successfully remove the trend, the dataset is said to be trend stationary.

The observations in a stationary time series are not dependent on time. Time series are stationary if they do not have trend or seasonal effects. 

When a time series is stationary, it can be easier to model. Statistical modeling methods assume or require the time series to be stationary to be effective.

### Augmented Dickey-Fuller test

```python
# calculate stationarity test of time series data
X = prices.values
result = adfuller(X)

print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')

for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

- The intuition behind this test is that it determines how strongly a time series is defined by a trend. 
- It uses an autoregressive model and optimizes an information criterion across multiple different lag values. 
- The null hypothesis of the test is that the time series is not stationary (has some time-dependent structure)

Here p-value > 0.05, we accept the null hypothesis and we conclude that the time series of our real-estate data are not stationary, that means they depends on time.

### Train-Test Split for Time-Series dataset

When training a model we need to make sure that we don’t use any information from the test set.

To avoid this, also called Data Leakage, we won't do:
- Cross-Validation on our temporal data
- Mix future and past data in the training and test sets

Linear Regression using Lags as features

### sqm_price : predict sqm_price of the next months (2016 to 2020)

```python
df_sqm_reg = pd.DataFrame(
    df_sqm['sqm_price'].values, index=df_sqm.month, columns=['sqm_price'])
df_sqm_reg = df_sqm_reg.groupby('month').mean()
```

```python
df_sqm_reg.head(3)
```

```python
# First lags and response: next
df_sqm_reg['lag1'] = df_sqm_reg['sqm_price'].shift(1)
df_sqm_reg['lag2'] = df_sqm_reg['sqm_price'].shift(2)

df_sqm_reg['next'] = df_sqm_reg['sqm_price'].shift(-1)
```

```python
# build lags
def buildLags(data, lagStart, lagEnd):
    lagColumns = list()
    for j in range(lagStart, lagEnd):
        colName = 'lag_'+str(j)
        data[colName] = data['sqm_price'].shift(j)
        lagColumns.append(colName)
    return data, lagColumns
```

```python
# Choose the number of lags = number of previous months to learn prices on (here 5 = 4 lags, 4 months)
fullFrame, lagColumns = buildLags(df_sqm_reg, 1, 5)
fullFrame.head(3)
```

```python
# drop nan, drop the n first rows (n = nb of lags chosen)
cleanFrame = fullFrame.dropna().reset_index(drop=True)

# create clean df: - response is next (the price of the next month)
#                  - features are the lags column that will be the features for the linear regression model

response = cleanFrame['next']
features = cleanFrame.loc[:, 'lag_1':]
features.head(3)
```

```python
# 90% of data for training set => test set starts from 2016 until 2020
i = int(len(features)*0.90)

X_train, y_train = features[:i], response[:i]
X_test, y_test = features[i:], response[i:]

reg = LinearRegression().fit(X_train, y_train)
print('Reg.score : ', reg.score(X_train, y_train))

mse = mean_squared_error(y_test, reg.predict(X_test), squared=False)
abs_err = mean_absolute_error(y_test, reg.predict(X_test))
print('RMSE: ', int(mse))
print('Abs error: %.2f' % abs_err)
```

By creating a dataset with lags as features, we could create a Linear Regression model to predict the sqm price of the next month.

With this method, we get an absolute error of $\$$36, less than the two previous methods ($\$$42 and $50)

```python
def create_newdf(response_record):

    # Create pandas Series from api response to specify type and add to the dataframe
    _id = pd.Series([response_record[i]['_id']
                     for i in range(len(response_record))], dtype='int')

    month = pd.Series([response_record[i]['month']
                       for i in range(len(response_record))], dtype='datetime64[ns]')

    town = pd.Series([response_record[i]['town']
                      for i in range(len(response_record))], dtype='string')

    # full address = block + street_name
    address = pd.Series([response_record[i]['block'] + ' ' + response_record[i]
                         ['street_name'] for i in range(len(response_record))], dtype='string')

    floor_area = pd.Series([response_record[i]['floor_area_sqm']
                            for i in range(len(response_record))], dtype='float')

    flat_type = pd.Series([response_record[i]['flat_type']
                           for i in range(len(response_record))], dtype='string')

    flat_model = pd.Series([response_record[i]['flat_model']
                            for i in range(len(response_record))], dtype='string')

    storey_range = pd.Series([response_record[i]['storey_range']
                              for i in range(len(response_record))], dtype='string')

    lease_commence_date = pd.Series([response_record[i]['lease_commence_date'] for i in range(
        len(response_record))], dtype='datetime64[ns]')

    remaining_lease = pd.Series([response_record[i]['remaining_lease']
                                 for i in range(len(response_record))], dtype='string')

    price = pd.Series([response_record[i]['resale_price']
                       for i in range(len(response_record))], dtype='float')

    # Create the Dataframe
    df = pd.DataFrame({'month': month, 'town': town, 'address': address, 'floor_area_sqm': floor_area,
                       'flat_type': flat_type, 'flat_model': flat_model, 'storey_range': storey_range, 'lease_commence_date': lease_commence_date,
                       'remaining_lease': remaining_lease,  'price': price})

    # Save in .csv
    df.to_csv(r'apidata.csv')
    return df
```

```python
def compute_dist(target, flat_data):

    dist_to_poi = []

    for h in range(len(flat_data)):
        print(h)
        # for each flat create a list of distance to poi
        dist_each_poi = []

        for t in range(len(target)):
            coords_1 = (target[0][t])
            coords_2 = (flat_data['LatLong'][h])
            dist = geopy.distance.geodesic(coords_1, coords_2).km

            dist_each_poi.append(dist)

        dist_to_poi.append(min(dist_each_poi))

    computed_dist = pd.Series(dist_to_poi, dtype='float64')

    return computed_dist
```
